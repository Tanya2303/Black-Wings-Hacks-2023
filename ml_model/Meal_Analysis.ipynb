# Data Preprocession

```python
import pandas as pd

left = pd.read_csv("/content/sample.csv")
right = pd.read_csv("/content/calories.csv")
```

```python
left = left.dropna(thresh=5, axis=1)
left
```

```python
left['Ingredients: Produce'] = left['Ingredients: Produce'].fillna(method='bfill')
left['Ingredients: Spices and Seasonings'] = left['Ingredients: Spices and Seasonings'].fillna(method='bfill')
left['Ingredients: Condiments'] = left['Ingredients: Condiments'].fillna(method='bfill')
left['Ingredients: Pasta and Rice'] = left['Ingredients: Pasta and Rice'].fillna(method='bfill')
left['Ingredients: Beverages'] = left['Ingredients: Beverages'].fillna(method='bfill')
left['Ingredients: Ethnic Foods'] = left['Ingredients: Ethnic Foods'].fillna(method='bfill')
left['Ingredients: Bakery/Bread'] = left['Ingredients: Bakery/Bread'].fillna(method='bfill')
left['Ingredients: Refrigerated'] = left['Ingredients: Refrigerated'].fillna(method='bfill')
left['Ingredients: Pantry'] = left['Ingredients: Pantry'].fillna(method='bfill')
left['Ingredients: Dairy'] = left['Ingredients: Dairy'].fillna(method='bfill')
```

```python
left['Ingredients'] = left.apply(lambda x: list([x['Ingredients: Produce'],x['Ingredients: Spices and Seasonings'],x['Ingredients: Condiments'],x['Ingredients: Pasta and Rice'],x['Ingredients: Beverages'],x['Ingredients: Ethnic Foods'],x['Ingredients: Bakery/Bread'],x['Ingredients: Refrigerated'],x['Ingredients: Pantry'],x['Ingredients: Dairy']]),axis=1)
```

```python
left = left.drop(["Ingredients: Produce","Ingredients: Spices and Seasonings","Ingredients: Condiments","Ingredients: Pasta and Rice","Ingredients: Beverages","Ingredients: Ethnic Foods","Ingredients: Bakery/Bread","Ingredients: Refrigerated","Ingredients: Pantry","Ingredients: Dairy","source","link"], axis=1)
```

```python
left['Ingredients'] = [x for x in left['Ingredients'] if x != 'nan']
```

```python
left = left.drop(["minutes_to_prepare","cuisine"], axis=1)
left.head(5)
```
